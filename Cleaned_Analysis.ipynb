```python
import re
import pandas as pd

def add_ota_classification_columns(dataframe, mapping_dict):
    # Define a custom function to handle both apostrophes and quotation marks
    def custom_contains(text, keyword):
        if pd.isna(text):
            return False
        pattern = re.compile(rf"\b(?:{keyword})(?:'s)?\b", re.IGNORECASE)
        return bool(pattern.search(text))

    # Create columns in the DataFrame for each category
    for classification in set(mapping_dict.values()):
        dataframe[classification] = 0

    # Iterate through each row and mark 1 in the corresponding columns
    for index, row in dataframe.iterrows():
        description = row['ICDDiagnosisCode_Desc']
        
        for keyword, classification in mapping_dict.items():
            if custom_contains(description, keyword):
                dataframe.loc[index, classification] = 1
    
    return dataframe
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2

def perform_chi2_feature_selection(dataframe, columns_to_keep, target_column, k='all'):
    # Drop rows with missing values in the specified columns
    cleaned_dataframe = dataframe.dropna(subset=columns_to_keep + [target_column])

    # Use the cleaned dataframe with selected columns
    cleaned_dataframe = cleaned_dataframe[columns_to_keep + [target_column]]

    # Separate the features (fracture locations) and the target column
    X = cleaned_dataframe.drop([target_column], axis=1)
    y = cleaned_dataframe[target_column]

    # Initialize SelectKBest with chi2 scoring function
    k_best = SelectKBest(score_func=chi2, k=k)

    # Fit and transform the features to get the scores and p-values
    X_new = k_best.fit_transform(X, y)
    p_values = k_best.pvalues_

    # Get the scores and corresponding feature names
    scores = k_best.scores_
    feature_names = X.columns

    # Create a DataFrame to display feature scores and p-values
    feature_scores_df = pd.DataFrame({'Feature': feature_names, 'Score': scores, 'P-Value': p_values})

    # Sort the DataFrame by score in descending order
    feature_scores_df = feature_scores_df.sort_values(by='Score', ascending=False)

    return feature_scores_df


```

```python
import pandas as pd
import scipy.stats as stats
from statistics import median, mode

def calculate_fasciotomy_statistics(dataframe, target_column, fasciotomy_column):
    # Filter the dataframe for people who had fasciotomies
    fasciotomy_1_df = dataframe[dataframe[fasciotomy_column] == 1]

    # Filter the dataframe for people who didn't have fasciotomies
    fasciotomy_0_df = dataframe[dataframe[fasciotomy_column] == 0]

    # Remove missing values
    fasciotomy_1_df = fasciotomy_1_df.dropna(subset=[target_column])
    fasciotomy_0_df = fasciotomy_0_df.dropna(subset=[target_column])

    # Calculate mean, median, and mode TOTALGCS for each group
    mean_totalgcs_fasciotomy_1 = fasciotomy_1_df[target_column].mean()
    mean_totalgcs_fasciotomy_0 = fasciotomy_0_df[target_column].mean()
    median_totalgcs_fasciotomy_1 = median(fasciotomy_1_df[target_column])
    median_totalgcs_fasciotomy_0 = median(fasciotomy_0_df[target_column])
    mode_totalgcs_fasciotomy_1 = mode(fasciotomy_1_df[target_column])
    mode_totalgcs_fasciotomy_0 = mode(fasciotomy_0_df[target_column])

    # Perform t-test to compare means
    t_statistic, p_value = stats.ttest_ind(fasciotomy_1_df[target_column], fasciotomy_0_df[target_column])

    # Print the statistics
    print("Statistics for people with fasciotomies:")
    print("Mean:", mean_totalgcs_fasciotomy_1)
    print("Median:", median_totalgcs_fasciotomy_1)
    print("Mode:", mode_totalgcs_fasciotomy_1)

    print("\nStatistics for people without fasciotomies:")
    print("Mean:", mean_totalgcs_fasciotomy_0)
    print("Median:", median_totalgcs_fasciotomy_0)
    print("Mode:", mode_totalgcs_fasciotomy_0)

    # Print the t-test results
    print("\nT-Statistic:", t_statistic)
    print("P-Value:", p_value)
```

```python
from scipy import stats
def check_categorical_dependency(crosstab_table, confidence_interval):
    stat, p, dof, expected = stats.chi2_contingency(crosstab_table)
    print ("Chi-Square Statistic value = {}".format(stat))
    print ("P - Value = {}".format(p))
    alpha = 1.0 - confidence_interval
    if p <= alpha:
        print('Dependent (reject H0)')
    else:
	      print('Independent (fail to reject H0)')
    return expected
     
```

```python
from itertools import combinations
import pandas as pd
all_combinations = []

for patient_idx, patient_row in just_fractures_fasciotomy_2017_2019.iterrows():
    patient_fracture_columns = [col for col in just_fractures_fasciotomy_2017_2019.columns if col not in ['INC_KEY', 'SEX', 'AGE']]
    patient_fractures = [col for col in patient_fracture_columns if patient_row[col] == 1]
    
    patient_combinations = list(combinations(patient_fractures, 2))
    all_combinations.extend(patient_combinations)

combinations_df = pd.DataFrame(all_combinations, columns=['Fracture1', 'Fracture2'])
combinations_counts = combinations_df.value_counts().reset_index()
combinations_counts.columns = ['Fracture1', 'Fracture2', 'Count']

combinations_counts = combinations_counts.sort_values(by='Count', ascending=False)

print(combinations_counts)
```

```python

unique_fractures = just_fractures_fasciotomy_2017_2019.columns
all_possible_combinations = list(combinations(unique_fractures, 2))
new_columns = {combo: [] for combo in all_possible_combinations}
for patient_idx, patient_row in fractures_fasciotomy_2017_2019_new.iterrows():
    patient_fractures = [col for col in unique_fractures if patient_row[col] == 1]
    
    for combo in all_possible_combinations:
        new_columns[combo].append(int(combo[0] in patient_fractures and combo[1] in patient_fractures))
new_columns_df = pd.DataFrame(new_columns)
final_df = pd.concat([fractures_fasciotomy_2017_2019_new, new_columns_df], axis=1)
print(final_df)
```
